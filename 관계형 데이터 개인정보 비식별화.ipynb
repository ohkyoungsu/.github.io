```python
sen = ["target", "thal"] #민감 정보라고 가정
```

```python
iden = ["sex", "cp"] #식별자라고 가정
```

```python
qu_id = ["age", "chol"] #준식별자라고 가정
```

```python
def chek_kAn(df, sen, id, qu_id):
    temp = df.copy()
    temp = temp.drop(id, axis = 1) #식별자 제거
    k_min = temp[qu_id + sen].value_counts().min() #준식별자를 기준으로 민감정보 K값 확인
    return temp, k_min
```

```python
temp, kAn = chek_kAn(df, sen, iden, qu_id)
```
